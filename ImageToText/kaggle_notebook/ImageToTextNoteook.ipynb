```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./stat-circle-workshop-test/pytesseact_demo/demo_code/images/002.png") #read an image
fig = plt.figure(figsize = [10,10])

plt.imshow(img)
```

```python
text = pytesseract.image_to_string(img) #extract text
print(text)
```

```python
file = open('output_perferct.txt','a') #write to a file
file.write(text)
file.close()
```

```python
img2 = cv2.imread("./stat-circle-workshop-test/pytesseact_demo/demo_code/images/006.jpg") #read an image
fig = plt.figure(figsize = [10,10])
plt.imshow(img2)
```

```python
text = pytesseract.image_to_string(img2)
print(text)
```

```python
img2= cv2.resize(img2,None, fx=.5, fy=0.5) #resizing the image #it's recommended if you’re working with images that have a DPI of less than 300 dpi (dots per inch)
print(img2.shape)
fig= plt.figure(figsize= [10,10])
plt.imshow(img2)
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY) #convert to gray scale
fig = plt.figure(figsize= [10,10])
plt.imshow(gray)
#plt.imshow(gray,cmap='gray', vmin=0, vmax=255)
```

```python
text2= pytesseract.image_to_string(gray)
print(text2)
```

```python
#some letters are coming from previous page.lets try adaptive thresholding
adaptive_threshold = cv2.adaptiveThreshold(gray,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY ,85, 11 )
fig = plt.figure(figsize= [10,10])
plt.imshow(adaptive_threshold)
#plt.imshow(adaptive_threshold,cmap='gray', vmin=0, vmax=255)
```

```python
text2= pytesseract.image_to_string(adaptive_threshold)
print(text2)
```
